```python
import pandas as pd
import numpy as np
df_b = pd.read_json('yelp_academic_dataset_business.json', lines=True)
```

```python
df_b
```

```python
df_b.dropna(subset=["attributes"])
df_b = df_b[df_b['attributes'].notnull()]
```

```python
list_attributes = set()
for i in df_b.attributes:
  temp = set(list(i.keys()))
  for j in temp:
    list_attributes.add(j)
print(list_attributes)
```

```python
len(list_attributes)
```

```python
count_attributes = dict()
for i in list_attributes:
  count_attributes[i] = 0
count_attributes
```

```python
for i in df_b.attributes:
  temp = set(list(i.keys()))
  for j in temp:
    count_attributes[j] += 1
print(count_attributes)
```

```python
for i in count_attributes:
  count_attributes[i] /= len(df_b)
count_attributes = dict(sorted(count_attributes.items(),key=lambda item:item[1], reverse=True))
print(count_attributes)
```

```python
frequent_attributes = []
for i in count_attributes:
  if count_attributes[i] > 0.3:
    frequent_attributes.append(i)
frequent_attributes
```

```python
def filter_frequent(row):
  lista = list(row.attributes.keys())
  for j in frequent_attributes:
    if j not in lista:
      return False
  return True
```

```python
df = df_b[df_b.apply(filter_frequent, axis=1)].reset_index()
df = df[["attributes", "stars"]]
df
```

```python
key = frequent_attributes[0]
def from_attribute(row):
  temp = row.attributes
  return temp[key]

for i in frequent_attributes:
  key = i
  df[i] = df.apply(from_attribute, axis=1)
df
```

```python
df.drop(['attributes'], axis = 1, inplace = True)
df
```

```python
regex = r"\'(.*?)\'"
df['WiFi'] =  df['WiFi'].str.extract(regex)
df['Alcohol'] =  df['Alcohol'].str.extract(regex)
df
```

```python
import ast
df = df[df['BusinessParking']!= "None"]

parking = df.BusinessParking.iloc[-1]
parking = ast.literal_eval(parking)
parking = list(parking.keys())

key = parking[0]
def from_parking(row):
  temp = row.BusinessParking
  temp = ast.literal_eval(temp)
  return temp[key]

for i in parking:
  key = i
  df[i] = df.apply(from_parking, axis=1)
df
```

```python
df.drop(['BusinessParking'], axis = 1, inplace = True)
df
```

```python
df = df[df['Ambience']!= "None"]

Ambience = df.Ambience.iloc[-1]
Ambience = ast.literal_eval(Ambience)
Ambience = list(Ambience.keys())
base = len(Ambience)

def filter_lenght(row):
  lista = ast.literal_eval(row.Ambience)
  lista = list(lista.keys())
  if (len(lista) == base):
    return True
  else:
    return False
df = df[df.apply(filter_lenght, axis=1)].reset_index()

key = Ambience[0]
def from_ambience(row):
  temp = row.Ambience
  temp = ast.literal_eval(temp)
  return temp[key]

for i in Ambience:
  key = i
  df[i] = df.apply(from_ambience, axis=1)
df
```

```python
df.drop(['Ambience', 'level_0', 'index'], axis = 1, inplace = True)
df
```
