```python
def suggest(user_id, a, data):
    friends=(a['user_id'][a['user_id1_y']==user_id])
    num=(a['business_id'][a['user_id1_y']==user_id])
    num=num.to_list()
    friends=friends.to_list()
    f=a['user_id1_y'][a['user_id']==user_id]
    num2=a['business_id'][a['user_id']==user_id]
    friends.extend(f.to_list())
    num.extend(num2.to_list())
    frame=pd.DataFrame()
    frame['friends']=friends
    frame['num_common']=num
    frame['num_common'].sort_values(ascending=True)
    frame=frame.head(1)
    lis=[]
    for p in frame['friends']:
        temp=data['business_id'][data['user_id1']==p]
        lis.extend(temp.to_list())
    lis=list(set(lis))
    l=data['business_id'][data['user_id1']==user_id]
    #sugg=[x for x in lis if x not in l]
    sugg=list(set(lis)-set(l))
    return sugg
   
```
