now we approach the 'space' data in another way, not considering states but considering cities, where we want to find the city with more users, the city with more reviews (which should be the same) and finally the city with a bigger rate reviews per user which could be actually the city where Yelp has become more popular

```python
num_pers=real_states.groupby(by=['state','city','user_id'])['rates'].count()
num_pers=num_pers.to_frame().reset_index()
num_pers['city']=num_pers['city'].str.lower()
num_pers
```

```python
num_pers_city=num_pers.groupby(by=['city','state'])['user_id'].count().to_frame().reset_index()
num_pers_city
```

```python
num_pers_city=num_pers_city[num_pers_city['user_id']>30000]
num_pers_city['city'][num_pers_city['state']=='WA']='vancouverWA'
num_pers_city
```

```python
fig, ax=plt.subplots(1,1,figsize=(20,10))
#ax.set_yscale('log')
sns.barplot(data=num_pers_city,x='city',y='user_id',palette='colorblind',ax=ax)
plt.ylabel('number_of_users')
plt.show()
```

the city with more users is Austin now if we count the number of reviews for this city we expect that it is also the city with most reviews

```python
num_rates_city=num_pers.groupby(by=['city','state'])['rates'].sum().to_frame().reset_index()
num_rates_city
```

```python
num_rates_city=num_rates_city[num_rates_city['rates']>50000]
num_rates_city['city'][num_rates_city['state']=='WA']='vancouverWA'
num_rates_city
```

```python
fig, ax=plt.subplots(1,1,figsize=(20,10))
#ax.set_yscale('log')
sns.barplot(data=num_rates_city,x='city',y='rates',palette='colorblind',ax=ax)
plt.ylabel('number_of_reviews')
plt.show()
```

as we expected the cities with a bigger number of users are also the cities with more reviews and also Austin is again the city with more reviews as we expected, now we want to find the city with highest rate review/user, to do so we merge the two dataset created for the plots

```python
rate_per_city=pd.merge(num_rates_city, num_pers_city)
rate_per_city
```

```python
rate_per_city['num_rev_us']=rate_per_city['rates']/rate_per_city['user_id']
rate_per_city
```

```python
fig, ax=plt.subplots(1,1,figsize=(20,10))
#ax.set_yscale('log')
sns.barplot(data=rate_per_city,x='city',y='num_rev_us',palette='colorblind',ax=ax)
plt.ylabel('number_of_reviews_per_user')
plt.show()
```

Because Austin is the city with the biggest number of review per user and also with the biggest number of users and of reviews is the city where Yelp is most used, although Austin is in TX and TX is not the state with the biggest number of reviews
